Credit Card Fraud detection

Importing Libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to a pandas dataframe
Credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
#First 5 rows of dataset
Credit_card_data.head()
```

```python
Credit_card_data.tail()
```

```python
Credit_card_data.info()
```

```python
#checking missing values in each column
Credit_card_data.isnull().sum()
```

```python
data=Credit_card_data.drop(Credit_card_data.tail(1).index)
```

```python
data.isnull().sum()
```

```python
# distribution of legit and fraudulent transaction
data['Class'].value_counts()
```

```python
# 0>> normal Transaction
# 1>> fraud transaction
```

```python
legit=data[data.Class==0]
fraud=data[data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transactions
data.groupby('Class').mean()
```

Under Sampling


Build a sample dataset containing distribution of normal transaction and Fraudulent Transaction

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.Class.value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
#Training data with Logistic Regression
model.fit(X_train,Y_train)
```

```python
X_train_pred=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,Y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_pred=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_pred,Y_test)
```

```python
print(test_data_accuracy)
```
